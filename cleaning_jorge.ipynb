```python
import numpy as np
import pandas as pd

df = pd.DataFrame(pd.read_csv("shark_attacks.csv"))

df.head()
```

```python
df.columns
```

```python
df = df.drop(columns = ['Date', 'Year', 'Type', 'Activity',
       'Name', 'Age', 'Injury', 'Fatal Y/N', 'Time', 'Species ',
       'Source', 'pdf', 'href formula', 'href', 'Case Number', 'Case Number.1',
       'original order', 'Unnamed: 21', 'Unnamed: 22'])
```

```python
df.columns = df.columns.str.lower()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["year"] = df["year"].fillna(0.0)
```

```python
df["year"] = df["year"].apply(lambda year: str(int(year)))
```

```python
df["full_date"] = df["date"] + " " +df["year"]
```

```python
df.dtypes
```

```python
df["full_date"] = df["full_date"].str.replace("-", " ", regex = False)
```

```python
df["full_date"] = df["full_date"].str.replace("st|nd|rd|th|Before|Reported|reported", "", regex = True)
```

```python
def remove_str(string):
    split = string.split()
    if len(split) >= 4:
        split = string.split()
        new_string = split[0] + " " + split[1] + " " + split[2]
        return new_string
    else: 
        return string
```

```python
df["full_date"] = df["full_date"].apply(remove_str)
```

```python
print(repr(df["full_date"][7000]))
```

```python
#df["final_date"] = pd.to_datetime(df["full_date"],format = "%d %B %Y", errors = "coerce")
```

```python
format1 = pd.to_datetime(df["full_date"], format="%d %B %Y", errors="coerce")
format2 = pd.to_datetime(df["full_date"], format="%Y %m %d", errors="coerce")
format3 = pd.to_datetime(df["full_date"], format="%d %b %Y", errors="coerce")
df["final_date"] = format1.fillna(format2).fillna(format3)
```
